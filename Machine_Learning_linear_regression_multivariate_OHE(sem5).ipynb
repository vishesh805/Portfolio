```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cars.csv')
```

```python
data
```

```python
dummies = pd.get_dummies(data['Car Model'])
dummies
```

```python
merge = pd.concat([data,dummies],axis='columns')
merge
```

```python
X= merge.drop(['Car Model','Sell Price($)'],axis = 'columns')
X
```

```python
Y = merge['Sell Price($)']
Y
```

```python
linearR = LinearRegression()
linearR.fit(X,Y)
```

```python
predicted_y=linearR.predict(X)
predicted_y
```

```python
linearR.score(X,predicted_y)
```
